## Evaluating Feature Imputation Strategies for Data Missing Completely at Random (MCAR)

The purpose of this project is to evaluate feature imputation strategies for imputing data missing completely at random (MCAR). 

The evaluted strategies include: 
1. Mean imputation 
2. Median imputation
3. KNN imputation
4. Iterative imputation

Using simulated data, these imputation strategies were evaluated in terms of bias in an imputed feature's mean, median, standard deviation, and correlation with another target feature. To test the robustness of these results, the effects of various factors (e.g., proportion of missing data) on the performance of the imputation strategies were further examined, averaged over 25 simulations. 

### Importing Libraries

```python
import pandas as pd 
import numpy as np 

from imputation_evaluation_functions import (
    get_data,
    get_mcar,
    get_imputations,
    plot_histograms,
    plot_scatterplots,
    get_imputation_results,
    iterated_missingness_simulation,
    gridsearch_missingness_simulation,
    missingness_simulation_plot,
    gridsearch_simulation_plot
)

np.set_printoptions(suppress=True)
```

### Evaluating Feature Imputation Strategies - Base Case

I will first evaluate a base case in which a simulated feature (<b>x1</b>) is missing 20% of data completely at random. The feature has a mean and median of approximately .00, a standard deviation of approximately 1.00, and a correlation with an external target (<b>y</b>) of approximately .50. 

For the purposes of the KNN/iterative imputation strategies, I also simulated a third feature (<b>x2</b>) that has approximate correlations of .50 with both <b>x1</b> and <b>y</b>.

Note: Many of these details will be systematically varied in subsequent simulations.

##### Generating Data

First, I will simulate the data such that all features have means/medians of .00, standard deviations of 1.00, and intercorrelations of .50.

```python
cmat = np.array([[1., .5, .5],
                 [.5, 1., .5],
                 [.5, .5, 1.]])

data = get_data(sample_size=10000, correlation_matrix=cmat, random_seed=1)

df = pd.DataFrame(data=data, columns=["x1", "x2", "y"])
```

Checking mean, median, and standard deviation.

```python
df.aggregate(["mean", "median", "std"])
```

Checking correlations between features.

```python
df.corr()
```

##### Feature Imputation for Data Missing Completely at Random (MCAR)

Next, I will create a new feature (<b>x1_na</b>) with 20% of the <b>x1</b> feature missing completely at random.

```python
df["x1_na"] = get_mcar(original_column=df["x1"], frac=.20, random_state=0)

df.isna().mean()
```

Next, I will use mean imputation (<b>x1_na_mean</b>), median imputation (<b>x1_na_median</b>), KNN imputation (<b>x1_na_knn</b>), and iterative imputation (<b>x1_na_iter</b>) to get four new <b>x1</b> features with the missing data imputed.

The <b>x2</b> feature is used for KNN/iterative imputation.

```python
imp_df = get_imputations(X=df[["x1_na", "x2"]], missing_column="x1_na")

df[imp_df.columns] = imp_df

df.head(3)
```

##### Visualizing Feature Imputation Results

Selecting the original <b>x1</b> feature and features reflecting various imputation strategies.

```python
x1_cols = df.columns[df.columns.str.startswith("x1")].tolist()
```

Examining the univariate distributions of these features.

```python
plot_histograms(data=df, columns=x1_cols)
```

Examining the bivariate distributions of these features with the target, <b>y</b>.

```python
plot_scatterplots(data=df, columns=x1_cols, target_column="y", alpha=.50)
```

Based on these visualizations, the mean and median imputation strategies have likely underestimated the variance of <b>x1</b> and attenuated the correlation between <b>x1</b> and <b>y</b>.

##### Descriptive Statistics and Errors Across Feature Imputation Strategies

Next, I will examine the descriptive statistics of the original <b>x1</b> feature and features reflecting various imputation strategies. Descriptive statistics include means, medians, standard deviations, and correlations with the target, <b>y</b>.

In addition, I will examine the extent to which the descriptive statistics of each of the features reflecting various imputation strategies deviate from the original <b>x1</b> feature.

```python
x1_missing_cols = df.columns[df.columns.str.startswith("x1_na_")].tolist()

descriptives, errors = get_imputation_results(
    data=df,
    original_column="x1",
    missing_columns=x1_missing_cols, 
    target_column="y"
)
```

```python
missingness_simulation_plot(descriptives)
```

```python
missingness_simulation_plot(errors)
```

All feature imputation strategies performed relatively well in terms of recovering the mean and median of the original feature.  

Mean/median imputation considerably underestimated the standard deviation of the original feature and attenuated the correlation between the original feature and the target, and  KNN/iterative imputation showed the same pattern but to a lesser degree.

### Evaluating Feature Imputation Strategies for MCAR Data - Base Case (25 Iterations)

To evaluate the robustness of the results, I will conduct the same simulation 25 times. Across the results for these iterations, I will take the average in descriptive statistics and error.

```python
na_sim_kwargs = {
    "data": df, 
    "original_column": "x1",
    "imputation_columns": ["x2"],
    "target_column": "y",
    "frac": .20,
}

descriptives, errors = iterated_missingness_simulation(k=25, **na_sim_kwargs)
```

```python
missingness_simulation_plot(descriptives)
```

```python
missingness_simulation_plot(errors)
```

The pattern of average error across the 25 iterations is consistent with the pattern of error obtained in the single base case.

### Evaluating Effect of Proportion of Data Missing on Feature Imputation Strategies (25 Iterations)

Next, I will systematically vary the proportion of missing data in <b>x1</b> to examine its effect on the performance of the feature imputation strategies, averaged over 25 iterations.

```python
prop_missing = np.arange(.10, .91, .20)

na_sim_kwargs = {
    "data": df, 
    "original_column": "x1",
    "imputation_columns": ["x2"],
    "target_column": "y"
}

descriptives, errors = gridsearch_missingness_simulation(
    parameter_name="frac",
    parameters=prop_missing,
    parameter_keys=prop_missing,
    k=25,
    **na_sim_kwargs
)
```

```python
gridsearch_simulation_plot(descriptives)
```

```python
gridsearch_simulation_plot(errors)
```

The overall pattern of average error is similar to that of the base case in that generally (1) all feature imputations strategies perform well in terms of recovering the mean/median of the original feature and (2) the mean/median imputation strategies result in greater underestimation of the original feature's standard deviation/correlation with the target as compared to the KNN/iterative imputation strategies.

With that said, the underestimation of the original feature's standard deviation/correlation with the target worsens as a function of the proportion of data missing across all feature imputation strategies. Moreover, this underestimation as a function of proportion of data missing is steeper for the mean/median imputation strategies as compared to the KNN/iterative imputation strategies.



### Evaluating the Effect of Correlation between x1 and x2 on Feature Imputation Strategies  (25 Iterations)

Next, I will systematically vary the correlation between <b>x1</b> and <b>x2</b> to examine its effect on the performance of various feature imputation strategies, averaged over 25 iterations.

Given that the KNN/iterative imputation strategies are the only strategies that rely on other predictor features, these changes should not affect the mean/median imputation strategies.

```python
correlations = np.arange(.10, .91, .20)

dfs = []
for rs, c in enumerate(correlations):
    cmat = np.array([[1., c, .5],
                     [c, 1., .5],
                     [.5, .5, 1.]])
    

    data = get_data(sample_size=10000, correlation_matrix=cmat, random_seed=rs)

    dfs.append(pd.DataFrame(data=data, columns=["x1", "x2", "y"]))
```

```python
na_sim_kwargs = { 
    "original_column": "x1",
    "imputation_columns": ["x2"],
    "target_column": "y",
    "frac": .20
}

descriptives, errors = gridsearch_missingness_simulation(
    parameter_name="data",
    parameters=dfs,
    parameter_keys=correlations,
    k=25,
    **na_sim_kwargs
)
```

```python
gridsearch_simulation_plot(descriptives)
```

```python
gridsearch_simulation_plot(errors)
```

Again, the overall pattern of average error is similar to that of the base case.

Expectedly, the performance of the mean/median imputation strategies was not affected by the correlation between <b>x1</b> and <b>x2</b>. 

In contrast, the underestimation of the original feature's standard deviation/correlation with the target improves (non-linearly) as a function of the correlation between <b>x1</b> and <b>x2</b> for the KNN/iterative imputation strategies. 

### Evaluating the Effect of Correlation between x2 and y on Feature Imputation Strategies  (25 Iterations)

Next, I will systematically vary the correlation between <b>x2</b> and <b>y</b> to examine its effect on the performance of various feature imputation strategies, averaged over 25 iterations.

Given that the KNN/iterative imputation strategies are the only strategies that rely on other predictor features, these changes should not affect the mean/median imputation strategies.

```python
correlations = np.arange(.10, .91, .20)

dfs = []
for rs, c in enumerate(correlations):
    cmat = np.array([[1., .5, .5],
                     [.5, 1., c],
                     [.5, c, 1.]])
    

    data = get_data(sample_size=10000, correlation_matrix=cmat, random_seed=rs)

    dfs.append(pd.DataFrame(data=data, columns=["x1", "x2", "y"]))
```

```python
na_sim_kwargs = { 
    "original_column": "x1",
    "imputation_columns": ["x2"],
    "target_column": "y",
    "frac": .20
}

descriptives, errors = gridsearch_missingness_simulation(
    parameter_name="data",
    parameters=dfs,
    parameter_keys=correlations,
    k=50,
    **na_sim_kwargs
)
```

```python
gridsearch_simulation_plot(descriptives)
```

```python
gridsearch_simulation_plot(errors)
```

Again, the overall pattern of average error is similar to that of the base case.

Expectedly, the performance of the mean/median imputation strategies was not affected by the correlation between <b>x2</b> and <b>y</b>. 

The correlation between <b>x2</b> and <b>y</b> only affected performance of the KNN/iterative imputation strategies with respect to the correlation between <b>x1</b> and <b>y</b>. The underestimation of the standard deviation of <b>x1</b> was not affected. Moreover, when <b>x2</b> and <b>y</b> are highly correlated, the correlation between <b>x1</b> and <b>y</b> is actually overestimated. 

Together, when using KNN/iterative imputation strategies, these results suggest that the correlation between <b>x2</b> and <b>y</b> can distort the correlation between <b>x1</b> and <b>y</b> to the extent that the correlation between <b>x2</b> and <b>y</b> deviates from the true correlation between <b>x1</b> and <b>y</b>.

### Summary

In general, across cases:
1. All feature imputations strategies perform well in terms of recovering the mean and median of the original feature
2. All feature imputation strategies tend to underestimate the standard deviation of the original feature and the correlation between the original feature and the target 
3. The mean and median imputation strategies result in greater underestimation of the standard deviation of the original feature and correlation between the original feature and the target as compared to the KNN/iterative imputation strategies

More specifically:
1. Proportion of data missing worsened the underestimation of the standard deviation of the original feature and correlation between the original feature and the target for all imputation strategies but more strongly for mean and median imputation strategies
2. Intercorrelations between the original feature, another predictor feature, and the target did not affect mean or median imputation strategies
3. Stronger correlation between the original feature and another predictor feature improved the underestimation of the standard deviation of the original feature and correlation between the original feature and the target for KNN and iterative imputation strategies
4. Stronger correlation between another predictor feature and the target did not improve underestimation of the standard deviation of the original feature for KNN and iterative imputation strategies
5. To the extent the correlation between another predictor feature and the target deviates from the correlation between the original feature and the target, then the correlation between the original feature and the target is distorted
